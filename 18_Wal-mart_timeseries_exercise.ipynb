### Walmart Sales Data

For the independent practice, we will analyze the weekly sales data from Walmart over a two year period from 2010 to 2012.

The data is again separated by store and by department, but we will focus on analyzing one store for simplicity.

The data includes:

- Store - the store number
- Dept - the department number
- Date - the week
- Weekly_Sales -  sales for the given department in the given store
- IsHoliday - whether the week is a special holiday week


```python
import pandas as pd
import numpy as np

%matplotlib inline
```

#### Loading the data and setting the DateTimeIndex

```python
data = pd.read_csv('./time_series_train.csv')
data['Date'] = pd.to_datetime(data['Date'])
data.set_index('Date', inplace=True)
data.head(100)
```

#### Filter the dataframe to Store 1 sales and aggregate over departments to compute the total sales per department.

```python
len(data)
```

```python
# TODO
df = data[data['Store']==1]
df.pivot_table(columns='Dept',index='Date',values='Weekly_Sales')
```

#### Plot the rolling_mean for `Weekly_Sales`. What general trends do you observe?

```python
df_week = df.groupby(by=df.index).sum()
```

```python
df_week[['Weekly_Sales']].resample('W').rolling(window=4).mean().plot()
```

#### Create an AR(1) model on the training data and compute the mean absolute error of the predictions.

```python
import statsmodels.api as sm
from sklearn.metrics import mean_absolute_error
```

```python
from statsmodels.graphics.tsaplots import plot_acf
plot_acf(df_week.Weekly_Sales, lags=10)
```

```python
# TODO
from statsmodels.tsa.arima_model import ARMA

model = ARMA(X, (1, 0)).fit()
model.summary() 
```

```python
prediction = model.predict('2012-03-02','2012-10-26')
```

```python
prediction.head()
```

```python
df_week[108:]['Weekly_Sales'].head()
```

```python
print(mean_absolute_error(prediction,df_week[108:]['Weekly_Sales']))
```

```python
import matplotlib.pyplot as plt

fig, ax = plt.subplots()
ax = df_week[108:]['Weekly_Sales'].plot(ax=ax)
# fig = prediction.plot(ax=ax)

fig = model.plot_predict(108,108+36, ax=ax, plot_insample=False)
```

#### Plot the residuals - where are their significant errors.

```python
# TODO
model.resid.plot()

# 3 times a year -> Holiday Seasons
```

```python
# TODO
plot_acf(model.resid, lags=50)
```
